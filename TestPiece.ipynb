```python
import pandas as pd

df = pd.read_csv("sensor_00212_data.csv")
roxbury_busdata = pd.read_csv("roxbury_busdata.csv")
roxbury_busdata["scheduled"] = roxbury_busdata["scheduled"].apply(lambda x: x[:10] + " " + x[11:-1])
roxbury_busdata["actual"] = roxbury_busdata["actual"].apply(lambda x: x[:10] + " " + x[11:-1])

sensor_lat = df["lat"][0]
sensor_long = df["lon"][0]

def bus_score(roxbury_busdata, x, x_lat, x_long):
    score = 0
    
    for index, row in roxbury_busdata.iterrows():
        if (int(row["actual"][5:7] + row["actual"][8:10]) > int(x[5:7] + x[8:10])):
            break;
        if (str(x[:-4]).__eq__(str(row["actual"][:-4]))) and (abs(int(x[14:16]) - int(row["actual"][14:16])) <= 5):
            score = (int(x[14:16]) - int(row["actual"][14:16])) * (int(x[14:16]) - int(row["actual"][14:16])) + (abs(row["lat"] - x_lat) + abs(row["long"] - x_long)) * 10000
    
    return score
```

```python
total = len(df)
progress = 0
df["bus_score"] = 0

for index, row in df.iterrows():
    x = row["date"]
    temp_bus_score = bus_score(roxbury_busdata, x, sensor_lat, sensor_long)
    row["bus_score"] = temp_bus_score
    
    # 'x' is a string representing a date in the format "YYYY-MM-DD"
    given_month_day = int(x[5:7] + x[8:10])

    # Create a mask for filtering rows based on the condition
    mask = given_month_day > (roxbury_busdata["actual"].str[5:7] + roxbury_busdata["actual"].str[8:10]).astype(int)

    # Apply the mask to filter the DataFrame
    roxbury_busdata = roxbury_busdata[~mask]

    progress += 1
    print("Bus Scores: ")
    print(temp_bus_score)
    print("Progress: ")
    print(progress/total)
```
